```python
import spacy
import os
import pandas as pd
import glob
from collections import defaultdict
import re

# Load English tokenizer, tagger, parser, NER and word vectors
nlp = spacy.load("en_core_web_sm")
```

```python
def cleanhtml(raw_html):
#   cleanr = re.compile('<.*?>')
    cleanr = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
    cleantext = re.sub(cleanr, '', raw_html)
    return cleantext
```

```python
def cleanEncoding(s):
    s= s.replace('Â\xa0', ' ')
    s= s.replace('\n',' ')
    return s
```

```python
def readFile(filename):
    text = open(filename, encoding="utf8").read()
    text = cleanhtml(cleanEncoding(text))
    return text
```

```python
root = "D:\\pari\\version 0.3\\en\\"
out = "D:/pari/out/"
```

```python
all_files = glob.glob( root + "/*.txt")

all_s = defaultdict(set)
for filename in all_files:
    text = readFile(filename)
    doc = nlp(text)
    
    # Find named entities, phrases and concepts
    for entity in doc.ents:
        if(entity.text not in all_s[entity.label_]):
            all_s[entity.label_].add(entity.text)
            f = open(out+entity.label_+".txt", "a", encoding="utf8")
            f.write(entity.text+"\n")
```

```python
all_files = glob.glob( root + "/*.txt")

all_s = defaultdict(set)
for filename in all_files:
    text = readFile(filename)
    doc = nlp(text)
    
    # Find named entities, phrases and concepts
    for token in doc:
        if(token.lemma_ not in all_s[token.pos_]):
            all_s[token.pos_].add(token.lemma_)
            if(token.pos_=='AUX'):
                fn = out+token.pos_+"_VERB.txt" #Windows does not allow "aux.txt"
            else:
                fn = out+token.pos_+".txt"
            f = open(fn, "a", encoding="utf8")
            f.write(token.lemma_+"\n")  
```

```python
root = "D:\\pari\\version 0.3\\"
out = "D:\\pari\\out\\"
dirs = glob.glob( root + "/*")

all_s = defaultdict(lambda: defaultdict(set))

for d in dirs:
    files = glob.glob (d + "/*.txt")
    lang = os.path.basename(d)
    for f in files:
        para = os.path.basename(f)
        all_s[para][lang] = readFile(f)

df = pd.DataFrame.from_dict(data=all_s, orient='index')
df.to_csv(out+'translations.csv', header=True, encoding='utf-8-sig')
```

```python
import spacy
from spacy_readability import Readability

read = Readability()
nlp.add_pipe(read, last=True)
```

```python
root = "D:\\pari\\out\\eng_full"
dirs = glob.glob( root + "/*")

g=0
r=0
d=0
c = 0

all_s = defaultdict(lambda: defaultdict(set))

for f in dirs:
    text = readFile(f)
    doc = nlp(text)
    g+=(doc._.flesch_kincaid_grade_level)
    r+=(doc._.flesch_kincaid_reading_ease)
    d+=(doc._.dale_chall)
    c+=1
    all_s[os.path.basename(f)]['FK Grade Level'] = doc._.flesch_kincaid_grade_level
    all_s[os.path.basename(f)]['FK Reading Ease'] = doc._.flesch_kincaid_reading_ease
    all_s[os.path.basename(f)]['Dale Chall'] = doc._.dale_chall
    
print(g/c)
print(r/c)
print(d/c)
df = pd.DataFrame.from_dict(data=all_s, orient='index')
df.to_csv(out+'readability.csv', header=True, encoding='utf-8-sig')
```
