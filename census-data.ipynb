```python
def ratio(col1, col2):
    t = df_no_total[df_no_total[col2]!=0]
    t['ratio'] = t[col1].apply(lambda x: float(x))/t[col2].apply(lambda x: float(x))
    t = t[t['ratio']!=0]
    t.sort_values('ratio', ascending=False)
    return t
```
